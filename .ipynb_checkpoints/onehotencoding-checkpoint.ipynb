```python
import pandas as pd
```

```python
#One hot encoder using pandas
```

```python
data=pd.read_csv('homeprices_.csv')
data
```

```python
dummies=pd.get_dummies(data['town'])
dummies=dummies.applymap(lambda x: 1 if x else 0)
```

```python
dummies
```

```python
merged=pd.concat([data,dummies],axis='columns')
merged
```

```python
final=merged.drop(['town'],axis='columns')
final
```

```python
x=final.drop(['west windsor','price'],axis='columns')
```

```python
x
```

```python
y=final['price']
y
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.score(x,y)
```

```python
model.predict([[3400,0,0]]) #34000 sqr ft home in west windsor
```

```python
#One hot encoder using sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
datale=data
datale
```

```python
datale['town']=le.fit_transform(datale['town'])
data
# monorore=0, robonsville=1, west windsor=2
```

```python
X=datale[['town','area']].values
X
```

```python
Y=data['price'].values
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer([('town',OneHotEncoder(),[0])],remainder='passthrough')
#columntransformer applies onehotencoder to the column which you specify through its name and column number. pass through means to remain all the other columns unchanged
```

```python
X=ct.fit_transform(X)
X
```

```python
X=X[:,1:]
X    
#because we are ignoring west windsor
```

```python
model.fit(X,Y)
```

```python
model.predict([[0,1,3400]])
```
