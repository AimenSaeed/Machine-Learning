```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('homeprices.csv')
data
```

```python
plt.scatter(data.area,data.price,color='red',marker='+')
plt.xlabel('Price(US$)')
plt.ylabel('Area(sq ft)')
```

```python
new_df=data.drop('price',axis='columns')
```

```python
new_df
```

```python
price=data.price
price
```

```python
#Creating Linear Regression Object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
#predicting the price of a house
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
area=pd.read_csv('areas.csv')
```

```python
area
```

```python
prices=reg.predict(area)
```

```python
area['predicted price']=prices
```

```python
area
```

```python
area.to_csv('Prediction.csv')
```
