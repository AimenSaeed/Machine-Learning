```python
#plotting the image
import matplotlib.pyplot as plt
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
#creating and training logistic regression model
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
model=LogisticRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(digits['data'],digits['target'],test_size=0.2)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
#confusion matrix
```

```python
y_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figure=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
x_train,x_test,y_train,y_test=train_test_split(flower.data,flower.target,test_size=0.2)
```

```python
flowermodel=LogisticRegression()
```

```python
flowermodel.fit(x_train,y_train)
```

```python
y_predicted=flowermodel.predict(x_test)
```

```python
y_test
```

```python
flowermodel.score(x_test,y_test)
```

```python
confusion=confusion_matrix(y_test,y_predicted)
```
