```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
lr=LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
rdg=Ridge(alpha=10)
rdg.fit(X,y)
print(rdg.coef_)
print(rdg.intercept_)
```

```python
rdg1=Ridge(alpha=100)
rdg1.fit(X,y)
print(rdg1.coef_)
print(rdg1.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),'g',label='alpha=0')
plt.plot(X,rdg.predict(X),'black',label='alpha=10')
plt.plot(X,rdg1.predict(X),'y',label='alpha=100')
plt.legend()
```

### Ridge Regression logic from scratch using OLS

```python
def ridge_regression(X,y,alpha):
    m=None
    b=None
    X_mean=np.mean(X)
    y_mean=np.mean(y)
    for i in range(X.shape[0]):
        num=np.sum(y[i]-y_mean)*(X[i]-X_mean)
        denum=np.sum((X[i]-X_mean)**2) 
        m= num/(denum+alpha)
        b=y_mean-(m*X_mean)
       
    return (f'm: {m}, b: {b}')
    
    
```

```python
rr=ridge_regression(X,y,10)
```
